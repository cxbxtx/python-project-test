# Прогнозирование оттока клиентов в сети отелей «Как в гостях»

## Описание проекта

### Общая информация

Заказчик этого исследования — сеть отелей «Как в гостях».

Чтобы привлечь клиентов, эта сеть отелей добавила на свой сайт возможность забронировать номер без предоплаты. Однако если клиент отменял бронирование, то компания терпела убытки. Сотрудники отеля могли, например, закупить продукты к приезду гостя или просто не успеть найти другого клиента.

Чтобы решить эту проблему, вам нужно разработать систему, которая предсказывает отказ от брони. Если модель покажет, что бронь будет отменена, то клиенту предлагается внести депозит. Размер депозита — 80% от стоимости номера за одни сутки и затрат на разовую уборку. Деньги будут списаны со счёта клиента, если он всё же отменит бронь.

<div class="alert alert-block alert-success">
<b>✔️ Комментарий ревьюера:</b> Цель работы понятна
</div>

### Исходные данные

Бизнес-метрика и другие данные
Основная бизнес-метрика для любой сети отелей — её прибыль. Прибыль отеля — это разница между стоимостью номера за все ночи и затраты на обслуживание: как при подготовке номера, так и при проживании постояльца.

В отеле есть несколько типов номеров. В зависимости от типа номера назначается стоимость за одну ночь. Есть также затраты на уборку. Если клиент снял номер надолго, то убираются каждые два дня.

Стоимость номеров отеля:

- категория A: за ночь — 1 000, разовое обслуживание — 400;
- категория B: за ночь — 800, разовое обслуживание — 350;
- категория C: за ночь — 600, разовое обслуживание — 350;
- категория D: за ночь — 550, разовое обслуживание — 150;
- категория E: за ночь — 500, разовое обслуживание — 150;
- категория F: за ночь — 450, разовое обслуживание — 150;
- категория G: за ночь — 350, разовое обслуживание — 150.

В ценовой политике отеля используются сезонные коэффициенты: весной и осенью цены повышаются на 20%, летом — на 40%.

На разработку системы прогнозирования заложен бюджет — 400 000. При этом необходимо учесть, что внедрение модели должно окупиться за год. Затраты на разработку должны быть меньше той выручки, которую система принесёт компании.

### Описание данных

В таблицах hotel_train и hotel_test содержатся одинаковые столбцы:
- `id` — номер записи;
- `adults` — количество взрослых постояльцев;
- `arrival_date_year` — год заезда;
- `arrival_date_month` — месяц заезда;
- `arrival_date_week_number` — неделя заезда;
- `arrival_date_day_of_month` — день заезда;
- `babies` — количество младенцев;
- `booking_changes` — количество изменений параметров заказа;
- `children` — количество детей от 3 до 14 лет;
- `country` — гражданство постояльца;
- `customer_type` — тип заказчика:
    - `Contract` — договор с юридическим лицом;
    - `Group` — групповой заезд;
    - `Transient` — не связано с договором или групповым заездом;
    - `Transient-party` — не связано с договором или групповым заездом, но связано с бронированием типа Transient.
- `days_in_waiting_list` — сколько дней заказ ожидал подтверждения;
- `distribution_channel` — канал дистрибуции заказа;
- `is_canceled` — отмена заказа;
- `is_repeated_guest` — признак того, что гость бронирует номер второй раз;
- `lead_time` — количество дней между датой бронирования и датой прибытия;
- `meal` — опции заказа:
    - `SC` — нет дополнительных опций;
    - `BB` — включён завтрак;
    - `HB` — включён завтрак и обед;
    - `FB` — включён завтрак, обед и ужин.
- `previous_bookings_not_canceled` — количество подтверждённых заказов у клиента;
- `previous_cancellations` — количество отменённых заказов у клиента;
- `required_car_parking_spaces` — необходимость места для автомобиля;
- `reserved_room_type` — тип забронированной комнаты;
- `stays_in_weekend_nights` — количество ночей в выходные дни;
- `stays_in_week_nights` — количество ночей в будние дни;
- `total_nights` — общее количество ночей;
- `total_of_special_requests` — количество специальных отметок.

## Формулировка ML-задачи на основе бизнес-задачи

Рассчитаем прибыль (убыток) по каждому клиенту - для этого создадим новый столбец, и рассчитаем общую прибыль до внедрения новой модели. Но для начала создадим словари:

```python
# cтоимость номера по каждой категории
prices_dict = {
    'A': 1000,
    'B': 800,
    'C': 600,
    'D': 550,
    'E': 500,
    'F': 450,
    'G': 350
}

# cтоимость разового обслуживания номера каждой категории
service_cost_dict = {
    'A': 400,
    'B': 350,
    'C': 350,
    'D': 150,
    'E': 150,
    'F': 150,
    'G': 150
}

# cезонный коэффициент: весной и осенью цены повышаются на 20%, летом — на 40%
seasonal_ratio_dict = {
    'January': 1,
    'February': 1,
    'March': 1.2,
    'April': 1.2,
    'May': 1.2,
    'June': 1.4,
    'July': 1.4,
    'August': 1.4,
    'September': 1.2,
    'October': 1.2,
    'November': 1.2,
    'December': 1
} 

# cезонный коэффициент с номерами месяцев
seasonal_ratio_num_dict = {
    1: 1,
    2: 1,
    3: 1.2,
    4: 1.2,
    5: 1.2,
    6: 1.4,
    7: 1.4,
    8: 1.4,
    9: 1.2,
    10: 1.2,
    11: 1.2,
    12: 1
}

# месяцы и их порядковые номера
months_dict = {
    'January': 1,
    'February': 2,
    'March': 3,
    'April': 4,
    'May': 5,
    'June': 6,
    'July': 7,
    'August': 8,
    'September': 9,
    'October': 10,
    'November': 11,
    'December': 12
}
```

С учетом созданных словарей создадим новые столбцы `prices`, `service_cost`, `seasonal_ratio`:

```python
hotel_full['prices'] = hotel_full['reserved_room_type'].apply(lambda x: prices_dict[x])

hotel_full['service_cost'] = hotel_full['reserved_room_type'].apply(lambda x: service_cost_dict[x])

hotel_full['seasonal_ratio'] = hotel_full['arrival_date_month'].apply(lambda x: seasonal_ratio_dict[x])

# проверим новые столбцы
hotel_full.sample(5)
```

Новые столбцы созданы. 

Создадим функцию, которая будет расчитывать количество ночей в каждом месяце:

```python
def nights_by_month_func(start_date, end_date):
    current_date = start_date
    result = [current_date]

    current_date = current_date.replace(day=1)
    while current_date <= end_date:
        current_date += timedelta(days=32)
        current_date = current_date.replace(day=1)
        result.append(datetime(current_date.year, current_date.month, 1).date())

    duration = []
    for current in result[:-1]:
        current_range = calendar.monthrange(current.year, current.month)
        current_duration = (current_range[1] - current.day) + 1

        if ((current.month == end_date.month) & (current.year == end_date.year)):
            duration.append(end_date.day)
        else:
            duration.append(current_duration)
            
    if len(duration) > 1:
        duration[-1] -= 1
    elif len(duration) == 1:
        duration[0] -= 1
        
    return duration
```

Используя полученную функцию, создадим новые столбцы в датасете:

```python
# cоздадим столбец с месяцем прибытия в формате номера
hotel_full['n_arrival_date_month'] = hotel_full['arrival_date_month'].apply(lambda x: months_dict[x])

# cоздадим столбец с датой прибытия
hotel_full['arrival_date'] = pd.to_datetime(dict(
    year=hotel_full.arrival_date_year, 
    month=hotel_full.n_arrival_date_month, 
    day=hotel_full.arrival_date_day_of_month)
)

# cоздадим столбец с датой выезда
hotel_full['departure_date'] = (
    hotel_full['arrival_date'] + pd.to_timedelta(hotel_full['total_nights'], unit='D')
)

# cоздадим столбец с количеством ночей в каждом месяце пребывания клиента
hotel_full['nights_by_month'] = hotel_full[['departure_date', 'arrival_date']].apply(
    lambda row: nights_by_month_func(row['arrival_date'], row['departure_date']) , axis=1
)

# проверим результат
hotel_full.head()
```

Столбцы созданы. Теперь создадим функцию для расчета выручки по каждому клиенту отеля:

```python
def revenue_func(df):
    revenue_list = []
    for index in df.index:
        if len(df.loc[index, 'nights_by_month']) == 2: 
            revenue = (
                (
                    df.loc[index, 'nights_by_month'][0] * 
                    df.loc[index, 'prices'] * 
                    df.loc[index, 'seasonal_ratio'] * 
                    (df.loc[index, 'is_canceled'] == 0)
                ) + 
                (
                    df.loc[index, 'nights_by_month'][1] * 
                    df.loc[index, 'prices'] * 
                    
                    seasonal_ratio_num_dict[((df.loc[index, 'n_arrival_date_month'] + 1) 
                              if df.loc[index, 'n_arrival_date_month'] < 12 
                              else 1)] * 
                    (df.loc[index, 'is_canceled'] == 0)
                )
            )

        else:
            revenue = (
                df.loc[index, 'prices'] * 
                df.loc[index, 'seasonal_ratio'] * 
                df.loc[index, 'total_nights'] * 
                (df.loc[index, 'is_canceled'] == 0)
            )
        revenue_list.append(revenue)

    return revenue_list
```

<div class="alert alert-block alert-warning">
<b>⚠️ Комментарий ревьюера: </b> Не очень хорошая практика использовать циклы в pandas. Всегда по-возможности старайся их избегать. Лучше использовать apply.
</div>

Создадим столбец `revenue` с информацией о выручке по каждому клиенту и проверим результат:

```python
hotel_full['revenue'] = revenue_func(hotel_full)
hotel_full.head()
```

<div class="alert alert-block alert-success">
<b>✔️ Комментарий ревьюера: </b> Да, верно
</div>

Рассчитаем прибыль по каждому клиенту, создав соответствующий столбец в датасете:

```python
# функция для расчета суммарных затрат на обслуживание
def total_costs_func(df):
    if df['is_canceled'] == 0:
        total_costs = (
            df['total_nights'] // 2 * df['service_cost'] + df['service_cost']
        )
    else:
        total_costs = df['service_cost']
    return total_costs

# столбец с суммарными затратами на обслуживание
hotel_full['total_costs'] = hotel_full.apply(total_costs_func, axis=1)

# столбец с прибылью по каждому клиенту
hotel_full['margin'] = hotel_full['revenue'] - hotel_full['total_costs']

# проверка результата
hotel_full.head()
```

Столбцы созданы. Теперь рассчитаем непосредственно саму прибыль отеля на тестовом наборе данных до внедрения системы прогнозирования:

```python
# создадим функцию для построения графика Precision-Recall
def precision_recall_plot_func(model=None, X=None, y=None, ax=None):
    probabilities_valid = model.predict_proba(X)
    precision, recall, thresholds = precision_recall_curve(y, probabilities_valid[:, 1])

    ax.step(recall, precision, where='post')
    ax.set_xlabel('Значение метрики Recall')
    ax.set_ylabel('Значение метрики Precision')
    ax.set_ylim([0.0, 1.05])
    ax.set_xlim([0.0, 1.0])
    ax.set_title('PR-кривая')
    ax.grid()
    
# Зададим функцию, которая построит график ROC_AUC
def roc_curve_plot_func(model=None, X=None, y=None, ax=None):
    probabilities_valid = model.predict_proba(X)
    fpr, tpr, thresholds = roc_curve(y, probabilities_valid[:, 1]) 
    
    ax.step(fpr, tpr)
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_ylim([0.0, 1.0])
    ax.set_xlim([0.0, 1.0])
    ax.set_title('ROC-кривая')
```
